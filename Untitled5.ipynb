```python
import pandas as pd
import numpy as np

df = pd.read_csv('us_county_sociohealth_data.csv')

column_name = 'population_density_per_sqmi'

mean = df[column_name].mean()
std = df[column_name].std()

threshold = 2

outliers = df[(np.abs(df[column_name] - mean) > threshold * std)]

print("Potential outliers:")
print(outliers)
```
